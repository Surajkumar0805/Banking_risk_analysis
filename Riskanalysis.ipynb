```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

```python
# HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(30,25))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#Insights of EDA:

1. Income and Financial Products
People with higher estimated income usually have slightly higher credit card balances, loans, and business lending. This means richer customers tend to use more financial products.

2. Credit Card Balance and Bank Loans
There is a strong positive link between credit card balance and bank loans. Customers who owe more on credit cards also tend to have larger bank loans.

3. Savings, Deposits, and Checking Accounts
These three accounts are moderately related. Customers who have a checking account often also keep money in savings or deposit accounts.

4. Foreign Currency Accounts
Foreign currency accounts are weakly related to other features. This suggests they are used by a special group of customers, maybe businesses or people who deal internationally.

5. Multicollinearity Check
None of the features show very high correlation (above 0.8). This means there is no strong overlap between the variables, and all can be used safely in further analysis or modeling.
```
